```python
nwalkers = 25
ndim = 1

def log_prob(params):
    x = params[0]
    return -0.5 * x * x  # log of Standard Normal

init_pos = 0.1 * np.random.randn(nwalkers, ndim)

sampler = emcee.EnsembleSampler(nwalkers,  ndim, log_prob)
t0 = time.time()
sampler.run_mcmc(init_pos, 5000, progress=True)

chain = sampler.get_chain()
t1 = time.time()

print(f"Generated {len(chain)} samples in {t1 - t0:.2f} seconds.")
```
